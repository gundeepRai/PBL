```python
import pandas as pd
import numpy as np
from sklearn import svm
from sklearn import model_selection
```

```python
df = pd.read_csv('datasetPBL(new).csv')
df=df.dropna()
df.head(10)

```

```python
df = df.drop('15', axis=1)
df.head(10)
```

```python
#dummy=pd.get_dummies(df,columns = 'date')
#print(dummy.head())

#df=pd.concat([df,dummy],axis=1)
#df.head(10)





df['timestamp'] = pd.to_datetime(df['timestamp'])

# Extract date information
df['year'] = df['timestamp'].dt.year
df['month'] = df['timestamp'].dt.month
df['day'] = df['timestamp'].dt.day

# Use get_dummies to create dummy variables
dummy = pd.get_dummies(df, columns=['year', 'month', 'day'])

# Print the DataFrame to see the updated columns

df.head()
```

```python
df = df.drop('timestamp', axis=1)
df = df.drop('date', axis=1)
df.head(10)
```

```python
X=df[['open','high','low','volume','marketCap','crypto_name','year','month','day']]
y=df['close']
```

```python
X_train,X_test,y_train,y_test=model_selection.train_test_split(X,y,test_size=.3,random_state=1)
```

```python
h1=svm.SVR()
```

```python
h1.fit(X_train, y_train)
```

```python
y_pred=h1.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import mean_absolute_error 
#from sklearn.metrics import 

mae = mean_absolute_error(y_test, y_pred)
print(f'Mean Absolute Error: {mae}')
print(mse(y_test,y_pred))

rmse = mse(y_test, y_pred, squared=False)
print(f'Root Mean Squared Error: {rmse}')
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(X_test, y_pred)
print(f'Accuracy: {accuracy}')
```
